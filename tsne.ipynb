```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
#merged_mat = np.load('./merged_mat.npy')
x = np.load('./mlp_4.7_testDataDict.npy', encoding = 'bytes').item()
```

```python
encoded = dict(x)[b'encoded']
label = dict(x)[b'y_true']
```

```python
label = label.astype(int)
```

```python
print(label.size)
print(encoded.size)

#print(np.shape(encoded))
```

```python
x
```

```python
model = TSNE(n_components = 2, random_state = 0)
x_tsne = model.fit_transform(encoded)
```

```python
#x_tsne.size

ind_1 = np.where(label == 1)
ind_0 = np.where(label == 0)
```

```python
fig = plt.figure(figsize=(12,12))

plt.scatter(x_tsne[:,0][ind_1], x_tsne[:,1][ind_1],c='r',marker='o',label='neuron',s=5)
plt.scatter(x_tsne[:,0][ind_0], x_tsne[:,1][ind_0],c='b',marker='s',label='tissue',s=3)
plt.legend(loc=2,fontsize=20)
plt.title('sklearn.TSNE',fontsize=20)

plt.show()

fig.savefig('sklearn_tsne.png')
```
